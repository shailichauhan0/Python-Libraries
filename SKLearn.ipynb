```python
#Value Imputation
#SimpleImputer includes three parameter (missing value(placeholder which has to be imputed), strategy(mean,mod,median),  fill value)
```

```python
from sklearn.impute import SimpleImputer
```

```python
df.age.isnull().sum()
```

```python
imp=SimpleImputer(strategy='mean')
df['age']=imp.fit_transform(df[['age']])
df.age.isnull().sum()
```

```python
def get_para(df):
    para = {}
    for col in df.columns[df.isnull().any()]:  # Fixed from `i` to `col`
        if df[col].dtype == 'float64' or df[col].dtype == 'int64' or df[col].dtype == 'int32':  # Fixed `col` usage
            strategy = 'mean'
        else:
            strategy = 'most_frequent'

        # Fixed `col` usage and ensured safe access to missing value
        missing_values = df[col][df[col].isnull()]
        if not missing_values.empty:
            missing_value = missing_values.values[0]
        else:
            missing_value = None

        para[col] = {"missing_value": missing_value, "strategy": strategy}  # Fixed from set to dict
    return para

get_para(df)
```

```python
para=get_para(df)
```

```python

for col, param in para.items():
    missing_values = param['missing_value']
    strategy = param['strategy']

    imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```
